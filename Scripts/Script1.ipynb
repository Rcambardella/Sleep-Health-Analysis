```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))

# Build the full path to the dataset inside the /data folder
data_path = os.path.join(project_root, "data", "Sleep_health_and_lifestyle_dataset.csv")

# Load the dataset
dataset = pd.read_csv(data_path)
dataset
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
# 219 Individuals out of 374 said they do not have sleeping disorders
```

```python
dataset['Sleep Disorder'].fillna('None', inplace=True)
dataset
```

```python
print(dataset['Stress Level'].value_counts())
```

```python
stress_sleep = dataset.groupby('Stress Level')['Sleep Duration'].mean().sort_index()
stress_sleep
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(8,5))
sns.lineplot(x=stress_sleep.index, y=stress_sleep.values)

plt.title('Average Sleep Duration by Stress Level', fontsize=14)
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('Average Sleep Duration (hours)', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

visuals_path = os.path.join(project_root, "visuals", "sleep_by_stress.png")

plt.savefig(visuals_path, bbox_inches='tight', dpi=300)

plt.show()
```

```python
activity_sleep = dataset.groupby('Physical Activity Level')['Sleep Duration'].mean().sort_index()
activity_sleep
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(8,5))
sns.lineplot(x=activity_sleep.index, y=activity_sleep.values)

plt.title('Average Sleep Duration by Physical Activity Level', fontsize=14)
plt.xlabel('Physical Activity Level', fontsize=12)
plt.ylabel('Average Sleep Duration (hours)', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

visuals_path = os.path.join(project_root, "visuals", "sleep_by_activity.png")

plt.savefig(visuals_path, bbox_inches='tight', dpi=300)

plt.show()
```

```python
quality_duration = dataset.groupby('Quality of Sleep')['Sleep Duration'].mean().sort_index()
quality_duration
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(8,5))
sns.lineplot(x=quality_duration.index, y=quality_duration.values)

plt.title('Average Sleep Duration by Quality of Sleep', fontsize=14)
plt.xlabel('Quality of Sleep', fontsize=12)
plt.ylabel('Average Sleep Duration (hours)', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()

visuals_path = os.path.join(project_root, "visuals", "sleep_by_quality.png")

plt.savefig(visuals_path, bbox_inches='tight', dpi=300)

plt.show()
```
